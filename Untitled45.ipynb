```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import metrics
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
#importing libraries
```

# Removing missing values

```python
#Completing Age column
df['Age'].fillna(df['Age'].mean(),inplace=True)
df
```

```python
#Completing Cabin column
number_of_elements=len(df["Cabin"])
print("Number of elements",number_of_elements)
print(df["Cabin"].value_counts())
df["Cabin"].fillna('G6',inplace=True)
df.tail()
```

```python
#Completing Embarked column
number_of_elements1=len(df["Embarked"])
print("Number of elements",number_of_elements1)
print(df["Embarked"].value_counts())
df["Embarked"].fillna('S',inplace=True)
df.tail()
```

```python
df.isnull().sum()
```

# Converting categorical features

```python
sex_nb={
    'male':0,
    'female':1
}
survived_nb={
    'No':0,
    'Yes':1
}
df.replace(sex_nb,inplace=True)
df.replace(survived_nb,inplace=True)
df
```

# Extracting titles from name column

```python
Title = {"Mrs": 1, "Mr": 2, "Miss": 3, "Major": 4, "Master": 5}
df['Title'] = df.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
#The expand=False flag returns a DataFrame
df = df.drop(['Name'], axis=1)
df
```

# Logistic Regression

```python
#features extraction
x = df[['PassengerId','Pclass','Age','Parch','Fare','Sex']]
y = df['Survived']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)  #splitting data with test size of 25%
logreg = LogisticRegression()   #build our logistic model
logreg.fit(x_train, y_train)  #fitting training data
y_pred  = logreg.predict(x_test)    #testing model’s performance
print("Accuracy={:.2f}".format(logreg.score(x_test, y_test)))
sns.regplot(x='Sex',y='Survived',data=df)#Most females survived and most males didn't survive
```

```python
sns.regplot(x='Age',y='Survived',data=df)#Most passengers with age between 2 and 28 are survived
```

```python
sns.regplot(x='Pclass',y='Survived',data=df)#people in the 1st and 2nd class survived the most 
```

# confusion Matrix

```python
confusion_matrix = pd.crosstab(y_test, y_pred, rownames=['Actual'], colnames=['Predicted'])
sns.heatmap(confusion_matrix, annot=True)
#True Positive=120
#False Negative=20
#True Negative=59
#False positive=19
#Accuracy= (TP+TN)/(TP+FP+FN+TN)=(120+59)/(120+19+20+59)=0,8211 : How many correct predictions have we made from all the observations
#Recall= (TP/(TP+FN))=(120/(120+20))=0,857 : Out of all the passengers that survived, how many did we get correct
#Precision= TP/(TP+FP)=(120/(120+19))=0,863 : Out of all the passengers that we predicted to be ‘Survived’, how many did we actually get right
```

# ROC

```python
#ROC :Receiver Operator Characteristic graphs provide a simple way to summarize all of the information
#Sensitivity: True positive rate :(TP/(TP+FN))
#The x-axis shows the False Positive Rate which is the same thing as Specificity=(FP/FP+TN)
# Plot ROC curve
from sklearn.metrics import roc_curve
# Generate ROC curve values: fpr, tpr, thresholds
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
# Plot ROC curve
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.show()
#fpr=FP/(TN+FP)
#tpr=Sensitivity=TP/TP+FN
```

# AUC

```python
#AUC :Area Under The Curve : makes it easy to compare one ROC curve to another 
#calculate AUC of model
auc = metrics.roc_auc_score(y_test, y_pred)
#print AUC score
print(auc)
#This model have predictions that they're  almost 80% correct 
```
